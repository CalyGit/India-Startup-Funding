```python
def outliers_IQR(dataframe):
    q1 = dataframe.quantile(0.25)
    q3 = dataframe.quantile(0.75)
    IQR = q3 - q1
    outliers = dataframe[(dataframe<(q1 - 1.5*IQR)) | (dataframe>(q3 + 1.5*IQR))]
    return outliers

outliers = outliers_IQR(complete_data_set["Amount($)"])
print(f"the number of outliers {str(len(outliers))}","\n")
number_amount = len(complete_data_set["Amount($)"])
print(f"Percentage outliers is {str(((len(outliers))/((number_amount))*100))}%", "\n")
print(f"The maximumum outlier is {str(outliers.max())}","\n")
print(f"The minimum outlier is {str(outliers.min())}")
```

### 3. Which sectors received the highest funding from 2018 to 2021?

```python
# Group the data by year and sector and calculate the total funding for each combination
yearly_sector_funding = df_clean.groupby(['Year_of_Funding', 'Sector'])['Amount'].sum().reset_index()

# Get the bottom ten sectors for each year
bottom_10_sectors_per_year = yearly_sector_funding.groupby('Year_of_Funding').apply(lambda x: x.nlargest(5, 'Amount')).reset_index(drop=True)

# Plot the bottom ten sectors per year
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 20), sharex=True)

for i, (year, ax) in enumerate(zip(range(2018, 2022), axes.flatten())):
    sectors = bottom_10_sectors_per_year[bottom_10_sectors_per_year['Year_of_Funding'] == year]
    sectors.sort_values('Amount', inplace=True)
    sectors.plot.barh(x='Sector', y='Amount', color='steelblue', ax=ax)
    ax.set_xlabel("Total Funding")
    ax.set_ylabel("Sector")
    ax.set_title(f"Top five Sectors in {year}")

    # Add data labels to the bar plot
    for j, v in enumerate(sectors['Amount']):
        ax.text(v + 0.5, j, str(v), color='black', va='center')

plt.tight_layout()
plt.show()
```

These are the sectors which have been successful in attracting investments and that indicate potential opportunities for entrepreneurs and investors.

### 4. Which sectors received the lowest funding per year?

```python
# Group the data by year and sector and calculate the total funding for each combination
yearly_sector_funding = df_clean.groupby(['Year_of_Funding', 'Sector'])['Amount'].sum().reset_index()

# Get the bottom ten sectors for each year
bottom_10_sectors_per_year = yearly_sector_funding.groupby('Year_of_Funding').apply(lambda x: x.nsmallest(5, 'Amount')).reset_index(drop=True)

# Plot the bottom ten sectors per year
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 20), sharex=True)

for i, (year, ax) in enumerate(zip(range(2018, 2022), axes.flatten())):
    sectors = bottom_10_sectors_per_year[bottom_10_sectors_per_year['Year_of_Funding'] == year]
    sectors.sort_values('Amount', inplace=True)
    sectors.plot.barh(x='Sector', y='Amount', color='steelblue', ax=ax)
    ax.set_xlabel("Total Funding")
    ax.set_ylabel("Sector")
    ax.set_title(f"Bottom five Sectors in {year}")

    # Add data labels to the bar plot
    for j, v in enumerate(sectors['Amount']):
        ax.text(v + 0.5, j, str(v), color='black', va='center')

plt.tight_layout()
plt.show()


```

From the dataset, quite a number of companies from different sectors had not disclosed their funding, which makes them to fall in the category of Sectors with the least funding. However, we can also note from 2019 that start-ups in the IoT, Sports and Legal sectors received the lowest funding of less than $200,000
